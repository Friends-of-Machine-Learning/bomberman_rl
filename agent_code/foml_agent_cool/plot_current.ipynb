```python
import matplotlib.pyplot as plt
import numpy as np

ACTIONS = ["UP", "RIGHT", "DOWN", "LEFT", "WAIT", "BOMB"]
```

```python
means = []
with open('means.txt', 'r') as file:
	for line in file:
		line = [float(num) for num in line.split(" ")]
		means.append(line)

rewards = []
with open('game_rewards.txt', 'r') as file:
	for line in file:
		rewards.append(float(line))
```

```python
means = np.array(means)
for i in range(6):
	plt.plot(means[:, i])
plt.legend(ACTIONS)
plt.show()
```

```python
plt.plot(rewards)
plt.show()
```
